```python
import os
import json
import pandas as pd
from pathlib import Path
from datetime import datetime
import shutil
import tempfile
from typing import Dict

# import singscore functions (adjust import if your package layout differs)
from singscore.singscore import *

def singscore_pathway_report(entrez_sample_path: str,
                             up_path: str,
                             down_path: str,
                             pathway_name: str,
                             norm_method: str = 'theoretical',
                             perms: int = 10000,
                             full_data: bool = True):
    """
    Run singscore for a pathway and return results in dict (JSON-able) format.

    Parameters
    ----------
    entrez_sample_path : str
        Path to expression matrix (tab-delimited). First column = gene id column name (e.g. "GeneName").
    up_path : str
        Path to file with up-regulated gene symbols (one per line).
    down_path : str
        Path to file with down-regulated gene symbols (one per line).
    pathway_name : str
        Display name for the pathway.
    norm_method : str
        Normalisation used by score(); default 'theoretical'.
    perms : int
        Number of permutations for empirical p-value (default 10000).
    full_data : bool
        Whether to request full_data=True from score().

    Returns
    -------
    dict
        JSON-serializable dictionary with pathway, gene_set_info, scores, statistics.
    """
    # --- Read expression data ---
    if not os.path.exists(entrez_sample_path):
        raise FileNotFoundError(f"Expression file not found: {entrez_sample_path}")
    # autodetect separator: assume tab-delimited like your example
    df = pd.read_csv(entrez_sample_path, sep="\t", header=0, index_col=0)
    # df index are gene identifiers (matching the signature)
    
    # --- Read gene lists ---
    def read_list(path):
        if not os.path.exists(path):
            raise FileNotFoundError(f"Gene list file not found: {path}")
        s = pd.read_csv(path, header=None, sep="\t", comment='#', engine='python')
        # first column contains gene symbols/ids
        return [str(x).strip() for x in s.iloc[:,0].tolist() if pd.notna(x)]
    
    up = read_list(up_path)
    down = read_list(down_path)
    
    # --- Check coverage in dataset (genes present) ---
    up_present = [g for g in up if g in df.index]
    down_present = [g for g in down if g in df.index]
    
    # --- Optional: warn about missing genes ---
    missing_up = set(up) - set(up_present)
    missing_down = set(down) - set(down_present)
    
    # --- Compute singscore ---
    # Use the original lists as inputs to score(); singscore internally ignores missing genes.
    scored = score(up_gene=up, down_gene=down, sample=df,
                   norm_method=norm_method, full_data=full_data)
    # scored is typically a DataFrame (rows = sample(s), columns include up_score, down_score, total_score)
    
    # Attempt to extract the numeric scores for the first sample (or the only sample)
    if isinstance(scored, pd.DataFrame):
        # If there are multiple rows (samples), pick the first row
        sample_row = scored.iloc[0]
        # Common column names in singscore outputs:
        # try several possibilities for column names
        col_names = {c.lower(): c for c in scored.columns}
        def pick_col(key_options):
            for k in key_options:
                if k in col_names:
                    return col_names[k]
            return None
        up_col = pick_col(['up_score', 's_up', 'sdir', 's_up_score', 's_up'])
        down_col = pick_col(['down_score', 's_down', 'sdir_down', 's_down_score', 's_down'])
        total_col = pick_col(['total_score', 's_total', 'stotal', 'score', 'total'])
        # fallback: if we can't find names, try to detect numeric columns by content
        if up_col is None or down_col is None or total_col is None:
            # find numeric columns and attempt to match by position or names containing 'up'/'down'
            numeric_cols = [c for c in scored.columns if pd.api.types.is_numeric_dtype(scored[c])]
            # name matching heuristic
            for c in scored.columns:
                cl = c.lower()
                if 'up' in cl and up_col is None:
                    up_col = c
                if 'down' in cl and down_col is None:
                    down_col = c
                if ('total' in cl or 'score' in cl) and total_col is None:
                    total_col = c
            # final fallback: first three numeric columns
            if (up_col is None or down_col is None or total_col is None) and len(numeric_cols) >= 3:
                up_col, down_col, total_col = numeric_cols[:3]
        # extract values (if still None, set to None)
        up_score = float(sample_row[up_col]) if up_col is not None else None
        down_score = float(sample_row[down_col]) if down_col is not None else None
        total_score = float(sample_row[total_col]) if total_col is not None else None
    else:
        # If scored is a scalar or other structure, attempt to convert
        # most robust approach is to convert to DataFrame first, but if not possible:
        raise ValueError("Unexpected return type from score(): expected DataFrame-like output.")
    
    # --- Permutation and empirical p-value ---
    # Use number of up/down genes actually present in data (so permutations are matched to effective sizes)
    n_up_used = len(up_present)
    n_down_used = len(down_present)
    permd = permutate(df, n_up=n_up_used, n_down=n_down_used, reps=perms, norm_method=norm_method)
    pvals_df = empiricalpval(permutations=permd, score=scored)
    # pvals_df often contains a column 'empirical p value' or similar; extract numeric value for first sample
    if isinstance(pvals_df, pd.DataFrame):
        # pick first numeric column
        numeric_cols = [c for c in pvals_df.columns if pd.api.types.is_numeric_dtype(pvals_df[c])]
        if len(numeric_cols) == 0:
            # try to find a column name containing 'p' or 'empir'
            numeric_cols = [c for c in pvals_df.columns if 'p' in c.lower() or 'empir' in c.lower()]
        if len(numeric_cols) == 0:
            empirical_p = None
        else:
            empirical_p = float(pvals_df.iloc[0][numeric_cols[0]])
    else:
        empirical_p = None
    
    # --- Build result dict ---
    result = {
        pathway_name: {
            "pathway_display_name": pathway_name,
            "gene_set_info": {
                "up_geneset_size": len(up),
                "down_geneset_size": len(down),
                "geneset_covered_in_sample": n_up_used + n_down_used
            },
            "scores": {
                "up_geneset_score": up_score,
                "down_geneset_score": down_score,
                "total_singscore": total_score
            },
            "statistics": {
                "p_value": empirical_p,
                "method": "permutation (empirical)",
                "permutations": perms
            }
        }
    }
    
    return result


def append_report_to_json(report: Dict, out_json_path: str,
                          overwrite: bool = True,
                          backup: bool = True):
    """
    Append or merge a `report` dict into a JSON file at out_json_path.

    Parameters
    ----------
    report : dict
        JSON-serializable dictionary containing one or more pathway entries
        (same format as your 'report' returned by singscore_pathway_report).
    out_json_path : str
        Path to the JSON file to update (will be created if missing).
    overwrite : bool
        If True, replace an existing pathway entry with the same key. If False,
        and the key exists, the function will create a new key by appending a suffix.
    backup : bool
        If True and the output file already exists, create a timestamped backup copy.
    """
    out_path = Path(out_json_path)
    # Load existing content if present
    if out_path.exists():
        try:
            with out_path.open("r", encoding="utf-8") as fh:
                existing = json.load(fh)
                if not isinstance(existing, dict):
                    raise ValueError("Existing JSON root is not a dict. Aborting.")
        except Exception as e:
            raise RuntimeError(f"Failed to read existing JSON file: {e}")

        # Optionally backup original file
        if backup:
            ts = datetime.now().strftime("%Y%m%d_%H%M%S")
            backup_path = out_path.with_name(out_path.stem + f"_backup_{ts}" + out_path.suffix)
            shutil.copy2(out_path, backup_path)
            print(f"Backup of existing JSON saved to: {backup_path}")
    else:
        existing = {}

    # Merge report into existing dictionary
    for pathway_key, pathway_val in report.items():
        if pathway_key in existing:
            if overwrite:
                existing[pathway_key] = pathway_val
                print(f"Overwrote existing entry for pathway: {pathway_key}")
            else:
                # Create unique key by appending timestamp (or a counter)
                ts = datetime.now().strftime("%Y%m%d_%H%M%S")
                new_key = f"{pathway_key}__dup__{ts}"
                existing[new_key] = pathway_val
                print(f"Pathway key exists. Added as new key: {new_key}")
        else:
            existing[pathway_key] = pathway_val
            print(f"Appended new pathway: {pathway_key}")

    # Write atomically to avoid partial writes (write to temp then move)
    tmp_fd, tmp_path = tempfile.mkstemp(dir=str(out_path.parent), prefix=".tmp_", suffix=".json")
    try:
        with open(tmp_fd, "w", encoding="utf-8") as tmpf:
            json.dump(existing, tmpf, indent=2, ensure_ascii=False)
            tmpf.flush()
        # Atomically replace
        Path(tmp_path).replace(out_path)
        print(f"Updated JSON written to: {out_path}")
    except Exception as e:
        # cleanup temp file on failure
        try:
            Path(tmp_path).unlink(missing_ok=True)
        except Exception:
            pass
        raise RuntimeError(f"Failed to write updated JSON file: {e}")
```

## Calculating the jaccard siliarity to get the top pathways for given drugs

```python
from typing import Dict, List, Iterable

def jaccard_similarity(list1: Iterable[str], list2: Iterable[str]) -> float:
    """
    Compute Jaccard similarity between two gene lists (case-insensitive).
    Returns 0.0 if union is empty.
    """
    s1 = {g.upper() for g in list1 if isinstance(g, str) and g.strip()}
    s2 = {g.upper() for g in list2 if isinstance(g, str) and g.strip()}
    if not s1 and not s2:
        return 0.0
    inter = s1.intersection(s2)
    union = s1.union(s2)
    return len(inter) / len(union) if union else 0.0

def compute_jaccard_for_df(
    gene_set_df: pd.DataFrame,
    drug_targets: Dict[str, List[str]],
    gene_col: str = "genes",
    setname_col: str = "set_name",
    genes_separator: str = ",",
    case_insensitive: bool = True,
) -> pd.DataFrame:
    """
    Compute Jaccard similarity between each gene set in gene_set_df and each drug in drug_targets.

    Parameters
    ----------
    gene_set_df : pd.DataFrame
        DataFrame with at least two columns: set_name and genes (comma-separated string OR list).
    drug_targets : dict
        Mapping from drug name -> list of target genes (strings).
    gene_col : str
        Column name in gene_set_df that contains gene lists (string or list).
    setname_col : str
        Column name that contains pathway / gene set names.
    genes_separator : str
        If gene_col contains strings, split by this separator to get genes.
    case_insensitive : bool
        If True, matching is case-insensitive (recommended).

    Returns
    -------
    pd.DataFrame
        Rows = gene sets; columns = set_name + one column per drug with Jaccard scores (float).
    """
    # Validate input
    if setname_col not in gene_set_df.columns or gene_col not in gene_set_df.columns:
        raise ValueError(f"DataFrame must contain columns '{setname_col}' and '{gene_col}'")

    # Prepare drug targets (normalize case if requested)
    if case_insensitive:
        drug_targets_norm = {drug: [g.upper() for g in genes if isinstance(g, str)] for drug, genes in drug_targets.items()}
    else:
        drug_targets_norm = {drug: [g for g in genes if isinstance(g, str)] for drug, genes in drug_targets.items()}

    results = []
    for idx, row in gene_set_df.iterrows():
        set_name = row[setname_col]
        raw_genes = row[gene_col]
        # If genes are a string (comma-separated), split them
        if isinstance(raw_genes, str):
            genes_list = [g.strip() for g in raw_genes.split(genes_separator) if g.strip()]
        elif isinstance(raw_genes, (list, tuple, set, pd.Series)):
            genes_list = [str(g).strip() for g in raw_genes if pd.notna(g)]
        else:
            genes_list = []
        # Normalize case if requested
        if case_insensitive:
            genes_list_norm = [g.upper() for g in genes_list]
        else:
            genes_list_norm = genes_list

        # Compute jaccard vs each drug
        row_result = {setname_col: set_name}
        for drug, targets in drug_targets_norm.items():
            score = jaccard_similarity(genes_list_norm, targets)
            row_result[drug] = score
        results.append(row_result)

    out_df = pd.DataFrame(results)
    # Optional: keep set_name as first column
    cols = [setname_col] + [c for c in out_df.columns if c != setname_col]
    out_df = out_df[cols]
    return out_df
```

### Saving relevant pathways by filteritng jaccard score > 0 for each drug

```python
def save_drug_relevant_pathways(drugname: str,
                                jaccard_df: pd.DataFrame,
                                gene_set_df: pd.DataFrame,
                                output_root: str = "singscore_new_code",
                                jaccard_col_check: bool = True,
                                save_json: bool = True) -> pd.DataFrame:
    """
    Filter gene_set_df for pathways relevant to a given drug using jaccard_df,
    save both a CSV and a JSON file.

    JSON format example:
    {
      "TNF-alpha Signaling via NF-kB": ["MARCKS", "IL23A", "NINJ1", ...],
      "MTOR_PATHWAY": ["G1", "G2"]
    }

    Parameters
    ----------
    drugname : str
        Drug name (must exist as a column in jaccard_df).
    jaccard_df : pd.DataFrame
        DataFrame containing 'set_name' and drug Jaccard scores.
    gene_set_df : pd.DataFrame
        DataFrame with columns ['set_name', 'genes'] (genes can be comma-separated string).
    output_root : str
        Folder where the subfolder per drug will be created.
    jaccard_col_check : bool
        Check column existence for safety.
    save_json : bool
        Whether to save a JSON version of the relevant pathways.

    Returns
    -------
    pd.DataFrame
        Filtered DataFrame containing relevant pathways.
    """


    # 2️⃣ Filter pathways with positive Jaccard similarity
    df1 = jaccard_df.sort_values(by=drugname, ascending=False)
    df2 = df1[df1[drugname] > 0]
    relevant_pathways = df2["set_name"].tolist()

    df4 = gene_set_df[gene_set_df["set_name"].isin(relevant_pathways)].reset_index(drop=True)

    # 3️⃣ Create output directory for this drug
    drug_dir = Path(output_root) / drugname.replace(" ", "_")
    drug_dir.mkdir(parents=True, exist_ok=True)

    # 5️⃣ Save JSON (key = set_name, value = list of genes)
    if save_json:
        json_dict = {}
        for _, row in df4.iterrows():
            genes = row["genes"]
            if isinstance(genes, str):
                gene_list = [g.strip() for g in genes.replace("\n", ",").split(",") if g.strip()]
            elif isinstance(genes, (list, tuple, set)):
                gene_list = list(genes)
            else:
                gene_list = []
            json_dict[row["set_name"]] = gene_list

        json_path = drug_dir / f"{drugname.replace(' ', '_')}_relevant_pathways.json"
        with open(json_path, "w", encoding="utf-8") as f:
            json.dump(json_dict, f, indent=2, ensure_ascii=False)

        print(f"✅ JSON saved to: {json_path}")

    return df4
```

```python
import os
import json
from pathlib import Path
from typing import List, Dict, Any, Optional
import pandas as pd

# Import singscore functions (adjust path if your package layout differs)
from singscore.singscore import *

def extract_single_numeric_from_df(df: pd.DataFrame, preferred_cols: Optional[List[str]] = None) -> Optional[float]:
    """
    Extract a single numeric value from a one-row DataFrame.
    pref_cols: list of column name substrings to try in order (case-insensitive)
    """
    if df is None:
        return None
    # if DataFrame has a single row and single numeric column, return it
    try:
        # If one row and one numeric column, return that
        if df.shape[0] == 1:
            numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
            # If preferred_cols provided, try to find match
            if preferred_cols:
                for pc in preferred_cols:
                    for c in df.columns:
                        if pc.lower() in c.lower() and (pd.api.types.is_numeric_dtype(df[c]) or pd.api.types.is_numeric_dtype(df.iloc[:, df.columns.get_loc(c)])):
                            return float(df.iloc[0][c])
            if numeric_cols:
                return float(df.iloc[0][numeric_cols[0]])
            # else try first column convertible to float
            first_val = df.iloc[0, 0]
            try:
                return float(first_val)
            except Exception:
                return None
        else:
            # more than one row: pick first row and try same logic
            first_row = df.iloc[0:1]
            numeric_cols = [c for c in first_row.columns if pd.api.types.is_numeric_dtype(first_row[c])]
            if numeric_cols:
                return float(first_row[numeric_cols[0]].iloc[0])
            first_val = first_row.iloc[0, 0]
            try:
                return float(first_val)
            except Exception:
                return None
    except Exception:
        return None

def compute_pathway_score_for_upgenes(
    data: pd.DataFrame,
    up_genes: List[str],
    norm_method: str = "theoretical",
    perms: int = 10000
) -> Dict[str, Any]:
    """
    Given expression DataFrame and up_genes list, compute up-only singscore and empirical p-value.
    Returns dict with gene_set_info, scores, statistics.
    """
    # Genes present
    up_present = [g for g in up_genes if g in data.index]
    n_up_used = len(up_present)

    # compute score (up only). down_gene=False per user snippet
    scored = score(up_gene=up_genes, down_gene=False, sample=data, norm_method=norm_method, full_data=False)
    total_score = extract_single_numeric_from_df(scored, preferred_cols=['total_score', 'total'])

    # permutations matched to n_up_used
    permd = permutate(data, n_up=n_up_used, n_down=False, reps=perms, norm_method=norm_method)
    pvals = empiricalpval(permutations=permd, score=scored)
    empirical_p = extract_single_numeric_from_df(pvals, preferred_cols=['empirical', 'p', 'empirical p'])

    result = {
        "gene_set_info": {
            "geneset_size": len(up_genes),
            "geneset_covered_in_sample": n_up_used
        },
        "scores": {
            "total_singscore": total_score
        },
        "statistics": {
            "p_value": empirical_p,
            "method": "permutation (empirical)",
            "permutations": perms
        }
    }
    return result

def run_singscore_on_drug_folder(
    drug_dir: Path,
    entrez_sample_path: str,
    norm_method: str = "theoretical",
    perms: int = 10000,
    output_root: Optional[Path] = None
) -> Dict[str, Any]:
    """
    Given a folder containing {drug}_relevant_pathways.json, read it and compute singscore for each pathway.
    Returns result dict for the drug and writes a JSON file <drug>_singscore_results.json in the drug folder (or output_root).
    """
    # discover json file
    drug_name = drug_dir.name
    in_json = None
    # accept "{drug}_relevant_pathways.json" or anything ending with "_relevant_pathways.json"
    for p in drug_dir.glob("*_relevant_pathways.json"):
        in_json = p
        break
    if in_json is None:
        # fallback: look for exact name
        candidate = drug_dir / f"{drug_name}_relevant_pathways.json"
        if candidate.exists():
            in_json = candidate
    if in_json is None:
        raise FileNotFoundError(f"No '*_relevant_pathways.json' found in {drug_dir}")

    # read expression data once
    if not Path(entrez_sample_path).exists():
        raise FileNotFoundError(f"Expression file not found: {entrez_sample_path}")
    df_expr = pd.read_csv(open(entrez_sample_path, 'r'), header='infer', sep='\t', index_col=0)
    # ensure index matches gene naming used in json (assumes 'GeneName' or similar; user used set_index(keys='GeneName'))
    # If index name is not 'GeneName' but first column is 'GeneName', above already set index_col=0.
    # Convert index to str for safety
    df_expr.index = df_expr.index.map(str)

    # read gene sets json
    with open(in_json, "r", encoding="utf-8") as f:
        pathway_dict = json.load(f)

    result_per_drug: Dict[str, Any] = {}
    for pathway_name, gene_list in pathway_dict.items():
        # gene_list expected to be list of gene symbols/ids
        # compute
        res = compute_pathway_score_for_upgenes(df_expr, gene_list, norm_method=norm_method, perms=perms)
        result_per_drug[pathway_name] = res

    # prepare output path
    if output_root is None:
        out_dir = drug_dir
    else:
        out_dir = Path(output_root) / drug_name
        out_dir.mkdir(parents=True, exist_ok=True)
    out_json = out_dir / f"{drug_name}_singscore_results.json"

    # wrap in top-level dict {Drug: {pathway...}}
    final_output = {drug_name: result_per_drug}
    with open(out_json, "w", encoding="utf-8") as f:
        json.dump(final_output, f, indent=2, ensure_ascii=False)
    print(f"Wrote results for '{drug_name}' to {out_json}")

    return final_output

def batch_run_on_drug_dirs(
    drug_dirs: List[str],
    entrez_sample_path: str,
    norm_method: str = "theoretical",
    perms: int = 10000,
    output_root: Optional[str] = None
) -> Dict[str, Any]:
    """
    Loop over a list of drug directory paths and run scoring for each.
    Returns a combined dict of all results.
    """
    combined = {}
    for d in drug_dirs:
        drug_dir = Path(d)
        if not drug_dir.exists() or not drug_dir.is_dir():
            print(f"Skipping non-directory: {d}")
            continue
        try:
            result = run_singscore_on_drug_folder(drug_dir, entrez_sample_path, norm_method=norm_method, perms=perms, output_root=output_root)
            combined.update(result)
        except Exception as e:
            print(f"Error processing {d}: {e}")
    return combined
```

```python
import os
import json
from pathlib import Path
from typing import List, Dict, Any, Optional
import pandas as pd
from tqdm import tqdm  # ✅ added for progress bars

# Import singscore functions (adjust path if needed)
from singscore.singscore import score, permutate, empiricalpval


def extract_single_numeric_from_df(df: pd.DataFrame, preferred_cols: Optional[List[str]] = None) -> Optional[float]:
    """Extract a single numeric value from a one-row DataFrame."""
    if df is None:
        return None
    try:
        if df.shape[0] == 1:
            numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
            if preferred_cols:
                for pc in preferred_cols:
                    for c in df.columns:
                        if pc.lower() in c.lower() and pd.api.types.is_numeric_dtype(df[c]):
                            return float(df.iloc[0][c])
            if numeric_cols:
                return float(df.iloc[0][numeric_cols[0]])
            return float(df.iloc[0, 0])
        else:
            first_row = df.iloc[0:1]
            numeric_cols = [c for c in first_row.columns if pd.api.types.is_numeric_dtype(first_row[c])]
            if numeric_cols:
                return float(first_row[numeric_cols[0]].iloc[0])
            return float(first_row.iloc[0, 0])
    except Exception:
        return None


def compute_pathway_score_for_upgenes(
    data: pd.DataFrame,
    up_genes: List[str],
    norm_method: str = "theoretical",
    perms: int = 5000
) -> Dict[str, Any]:
    """
    Given expression DataFrame and up_genes list, compute up-only singScore and empirical p-value.
    """
    up_present = [g for g in up_genes if g in data.index]
    n_up_used = len(up_present)

    # Calculate singscore (up only)
    scored = score(up_gene=up_genes, down_gene=False, sample=data, norm_method=norm_method, full_data=False)
    total_score = extract_single_numeric_from_df(scored, preferred_cols=['total_score', 'total'])

    # Calculate permutation-based p-value
    permd = permutate(data, n_up=n_up_used, n_down=False, reps=perms, norm_method=norm_method)
    pvals = empiricalpval(permutations=permd, score=scored)
    empirical_p = extract_single_numeric_from_df(pvals, preferred_cols=['empirical', 'p', 'empirical p'])

    result = {
        "gene_set_info": {
            "geneset_size": len(up_genes),
            "geneset_covered_in_sample": n_up_used
        },
        "scores": {
            "total_singscore": total_score
        },
        "statistics": {
            "p_value": empirical_p,
            "method": "permutation (empirical)",
            "permutations": perms
        }
    }
    return result


def run_singscore_on_drug_folder(
    drug_dir: Path,
    entrez_sample_path: str,
    norm_method: str = "theoretical",
    perms: int = 5000,
    output_root: Optional[Path] = None
) -> Dict[str, Any]:
    """
    Given a folder containing {drug}_relevant_pathways.json, read it and compute singScore for each pathway.
    """
    drug_name = drug_dir.name
    # Find input JSON file
    in_json = None
    for p in drug_dir.glob("*_relevant_pathways.json"):
        in_json = p
        break
    if in_json is None:
        candidate = drug_dir / f"{drug_name}_relevant_pathways.json"
        if candidate.exists():
            in_json = candidate
    if in_json is None:
        raise FileNotFoundError(f"No '*_relevant_pathways.json' found in {drug_dir}")

    # Load expression data
    if not Path(entrez_sample_path).exists():
        raise FileNotFoundError(f"Expression file not found: {entrez_sample_path}")
    df_expr = pd.read_csv(open(entrez_sample_path, 'r'), header='infer', sep='\t', index_col=0)
    df_expr.index = df_expr.index.map(str)

    # Load pathways JSON
    with open(in_json, "r", encoding="utf-8") as f:
        pathway_dict = json.load(f)

    result_per_drug: Dict[str, Any] = {}

    # ✅ tqdm progress bar for pathways
    for pathway_name, gene_list in tqdm(pathway_dict.items(), desc=f"Processing pathways for {drug_name}", leave=False):
        try:
            res = compute_pathway_score_for_upgenes(df_expr, gene_list, norm_method=norm_method, perms=perms)
            result_per_drug[pathway_name] = res
        except Exception as e:
            print(f"⚠️ Error computing {pathway_name} for {drug_name}: {e}")

    # Prepare output
    if output_root is None:
        out_dir = drug_dir
    else:
        out_dir = Path(output_root) / drug_name
        out_dir.mkdir(parents=True, exist_ok=True)

    out_json = out_dir / f"{drug_name}_singscore_results.json"

    final_output = {drug_name: result_per_drug}
    with open(out_json, "w", encoding="utf-8") as f:
        json.dump(final_output, f, indent=2, ensure_ascii=False)

    print(f"✅ Wrote results for '{drug_name}' → {out_json}")
    return final_output


def batch_run_on_drug_dirs(
    drug_dirs: List[str],
    entrez_sample_path: str,
    norm_method: str = "theoretical",
    perms: int = 5000,
    output_root: Optional[str] = None
) -> Dict[str, Any]:
    """
    Loop over a list of drug directories and run singScore for each.
    """
    combined = {}
    # ✅ tqdm for drugs
    for d in tqdm(drug_dirs, desc="Processing drugs"):
        drug_dir = Path(d)
        if not drug_dir.exists() or not drug_dir.is_dir():
            print(f"⚠️ Skipping non-directory: {d}")
            continue
        try:
            result = run_singscore_on_drug_folder(
                drug_dir,
                entrez_sample_path,
                norm_method=norm_method,
                perms=perms,
                output_root=output_root
            )
            combined.update(result)
        except Exception as e:
            print(f"❌ Error processing {d}: {e}")
    return combined
```

```python
#!/usr/bin/env python3
import json
from pathlib import Path
from typing import Optional
import pandas as pd
from tqdm import tqdm
# make sure singscore is importable in your env
from singscore.singscore import score, permutate, empiricalpval

# --------- Configuration (edit as needed) ----------
DRUG_PARENT = Path("singscore_new_code")   # parent folder containing drug subfolders
ENTREZ_SAMPLE_PATH = DRUG_PARENT / "UK65_clean_gene_log_tpm.txt"
PERMS = 10000
N_DRUGS = None  # set to int to limit for debugging, or None to run all
OUTPUT_ROOT: Optional[Path] = None  # if None, saves into each drug folder
# ---------------------------------------------------

def _extract_number(df: pd.DataFrame, preferred: tuple = ("total", "empir", "p")) -> Optional[float]:
    """
    Small helper: return a single numeric from a one-row DataFrame.
    Tries preferred substrings in column names, else first numeric column, else first cell castable to float.
    """
    if df is None or df.shape[0] == 0:
        return None
    row = df.iloc[0]
    # try preferred substrings
    for sub in preferred:
        for col in df.columns:
            if sub in col.lower():
                try:
                    return float(row[col])
                except Exception:
                    continue
    # numeric dtype fallback
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            try:
                return float(row[col])
            except Exception:
                continue
    # last resort try cast first value
    try:
        return float(row.iloc[0])
    except Exception:
        return None

# ---- load expression once ----
if not ENTREZ_SAMPLE_PATH.exists():
    raise FileNotFoundError(f"Expression file not found: {ENTREZ_SAMPLE_PATH}")
expr = pd.read_csv(ENTREZ_SAMPLE_PATH, sep="\t", header=0, index_col=0)
expr.index = expr.index.map(str)  # ensure strings

# ---- discover drug folders ----
drug_dirs = [p for p in DRUG_PARENT.iterdir() if p.is_dir()]
if N_DRUGS:
    drug_dirs = drug_dirs[:N_DRUGS]

# ---- batch run with progress bars ----
all_results = {}
for drug_dir in tqdm(drug_dirs, desc="Drugs"):
    drug_name = drug_dir.name
    # find relevant json file
    js = None
    for p in drug_dir.glob("*_relevant_pathways.json"):
        js = p
        break
    if js is None:
        fallback = drug_dir / f"{drug_name}_relevant_pathways.json"
        if fallback.exists():
            js = fallback
    if js is None:
        tqdm.write(f"Skipping {drug_name}: no *_relevant_pathways.json found")
        continue

    try:
        pathway_dict = json.loads(js.read_text(encoding="utf-8"))
    except Exception as e:
        tqdm.write(f"Failed to read {js}: {e}")
        continue

    per_drug_res = {}
    # pathway-level progress
    for pathway_name, gene_list in tqdm(list(pathway_dict.items()),
                                        desc=f"{drug_name} pathways",
                                        leave=False):
        try:
            # ensure gene_list is list
            if isinstance(gene_list, str):
                genes = [g.strip() for g in gene_list.split(",") if g.strip()]
            else:
                genes = [str(g).strip() for g in gene_list]

            # genes present in data
            up_present = [g for g in genes if g in expr.index]
            n_up_used = len(up_present)

            # score (up-only)
            scored = score(up_gene=genes, down_gene=False, sample=expr,
                           norm_method='theoretical', full_data=False)
            total_score = _extract_number(scored, preferred=("total",))

            # permutations and empirical p-value (match size to genes found)
            permd = permutate(expr, n_up=n_up_used, n_down=False, reps=PERMS, norm_method='theoretical')
            pvals = empiricalpval(permutations=permd, score=scored)
            empirical_p = _extract_number(pvals, preferred=("empir","p"))

            per_drug_res[pathway_name] = {
                "gene_set_info": {
                    "geneset_size": len(genes),
                    "geneset_covered_in_sample": n_up_used
                },
                "scores": {
                    "total_singscore": total_score
                },
                "statistics": {
                    "p_value": empirical_p,
                    "method": "permutation (empirical)",
                    "permutations": PERMS
                }
            }
        except Exception as e:
            tqdm.write(f"Error processing {drug_name} -> {pathway_name}: {e}")
            # record failure with None values
            per_drug_res[pathway_name] = {
                "gene_set_info": {"geneset_size": len(gene_list) if hasattr(gene_list, '__len__') else None,
                                  "geneset_covered_in_sample": None},
                "scores": {"total_singscore": None},
                "statistics": {"p_value": None, "method": "permutation (empirical)", "permutations": PERMS},
                "error": str(e)
            }

    # write per-drug json
    out_dir = (Path(OUTPUT_ROOT)/drug_name) if OUTPUT_ROOT else drug_dir
    out_dir.mkdir(parents=True, exist_ok=True)
    out_path = out_dir / f"{drug_name}_singscore_results.json"
    out_json = {drug_name: per_drug_res}
    out_path.write_text(json.dumps(out_json, indent=2, ensure_ascii=False), encoding="utf-8")
    all_results[drug_name] = per_drug_res
    tqdm.write(f"Wrote {out_path}")

# optional: save combined summary
combined_path = DRUG_PARENT / "all_drugs_singscore_summary.json"
combined_path.write_text(json.dumps(all_results, indent=2, ensure_ascii=False), encoding="utf-8")
print("Done. Summary saved to", combined_path)
```
